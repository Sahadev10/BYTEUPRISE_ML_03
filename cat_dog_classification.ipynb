```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle

dir = '/home/sahadev/Desktop/notebook/cats_dogs/archive/training_set/training_set'

categories = ['cats','dogs']

data = []

for category in categories:

    path = os.path.join(dir,category)
    label=categories.index(category)

    for img in os.listdir(path):

        imgpath = os.path.join(path,img)
        pet_img=cv2.imread(imgpath,0)
        try:
            
            pet_img = cv2.resize(pet_img, (50, 50))  # Store the resized image back
            image = np.array(pet_img).flatten()

            data.append([image,label])
        except Exception as e:
            pass
print(len(data))
pick_in = open('data1.pickle','wb')
pickle.dump(data,pick_in)
pick_in.close()
             
```

```python
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC


pick_in = open('data1.pickle','rb')
data = pickle.load(pick_in)
pick_in.close()

random.shuffle(data)
features = []
labels = []

for feature, label in data:
    features.append(feature)
    labels.append(label)
    


xtrain,xtest,ytrain,ytest = train_test_split(features,labels,test_size=0.25)


model = SVC(C=1,kernel='poly',gamma='auto')
model.fit(xtrain,ytrain)
```

```python

pick = open('model.sav','wb')
pickle.dump(model,pick)
pick.close()
```

```python
import pickle

pick = open('model.sav','rb')
model = pickle.load(pick)
pick.close()
```

```python
import cv2
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC


prediction = model.predict(xtest)
accuracy = model.score(xtest, ytest)

categories = ['cats','dogs']

print('accuracy :',accuracy)

print('prdiction is :',categories[prediction[0]])

my_pet=xtest[0].reshape(50,50)
# plt.imshow(my_pet,cmap='gray')

cv2.imshow('image',my_pet)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from sklearn.svm import SVC


# Assuming `image_path` is the path to the image you want to predict
image_path = '/home/sahadev/Desktop/notebook/cats_dogs/archive/ca.jpg'

# Read the image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Resize the image to match the input size expected by the model
image = cv2.resize(image, (50, 50))

# Flatten the image
image_flattened = image.flatten()

# Reshape the flattened image to match the input shape expected by the model
image_reshaped = image_flattened.reshape(1, -1)

# Perform prediction
prediction = model.predict(image_reshaped)

# Assuming `categories` is defined earlier in your code
categories = ['cats', 'dogs']

# Print prediction
print('Prediction is:', categories[prediction[0]])

# Display the image
cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
